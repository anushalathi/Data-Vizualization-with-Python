```python
days = [1,2,3,4,5,6,7]
sales_1 = [160, 150, 140, 145, 175, 165, 180] #company 1
sales_2 = [70, 90, 160, 150, 140, 145, 175] #company 2
plt.figure(figsize=(4,2),dpi=140)
plt.plot(days,sales_1)
plt.plot(days,sales_2)

plt.xlabel("Days of week")
plt.ylabel("Sales")
plt.grid()
plt.show()
```
